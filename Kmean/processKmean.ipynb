```python
def analyst(content):
    tmp = str(content)
    a = len(tmp.split("."))-1
    # maxLen1 = max(maxLen,len(tmp.split("."))-1)
    return a
```

```python
lengh1 = []
lengh2 = []
maxLen1 = 0


for x in range(data.shape[0]):
    if x% 1000==0:
        print(x)
    tmp = str(data.iloc[x]['headline'])
    a = analyst(tmp)
    lengh1.append(a)
    lengh2.append(analyst(data.iloc[x]['text']))



```
